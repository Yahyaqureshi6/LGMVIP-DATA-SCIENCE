Image to pencil sketch

```python
import cv2

from IPython.display import Image

from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
image = cv2.imread('horse.jpg')
cv2_imshow(image)
```

converted to gray scale



```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```


Invert the new gray scale

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

Blur image

```python
blur_image = cv2.GaussianBlur(inverted_image,(21,21),0)
cv2_imshow(blur_image)
```

Invert blur image

```python
inverted_blur_image = 255 - blur_image
cv2_imshow(inverted_blur_image)
```

Pencil sketch

```python
pencil_sketch = cv2.divide(gray_image, inverted_blur_image, scale=256.0)
cv2_imshow(pencil_sketch)
```


Displaying Both Original and Pencil **Sketch**

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
