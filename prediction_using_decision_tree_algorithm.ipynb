creating the model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score

a = df.iloc[:, 1:-1].values
b = df.iloc[:, -1].values
```

```python
#Split data in training and test set

a_train, a_test, b_train, b_test = train_test_split(a, b, test_size = 0.2, random_state = 42)
```

```python
# Defining the decision tree algo

from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(a_train, b_train)
```

```python
#Evaluating the model before training on the entire Train set

from sklearn.metrics import accuracy_score
b_prediction = dtree.predict(a_test)
print("Accuracy: ", accuracy_score(b_test, b_prediction) * 100, '%')
```

```python
# Training model on entire train set
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(a,b)
```

```python
from sklearn.tree import plot_tree

featName=['sepal length','sepal width','petal length','petal width']
clsName=['Iris-Setosa','Iris-Versicolor','Iris-virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8,8), dpi = 500)

output = plot_tree(dtree, feature_names = featName, class_names = clsName, filled = True);
for j in output:
    arrow = j.arrow_patch
    if arrow is not None:
        arrow.set_edgecolor('green')
        arrow.set_linewidth(1.5)
fig.savefig('image.png')
```
